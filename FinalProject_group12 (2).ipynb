## Background and Prior Work

Marriage and divorce rates in the United States have long been subjects of interest, not only for sociologists and demographers but also for those in the entertainment industry. The portrayal of romantic relationships in films, particularly romantic comedies, often reflects societal norms, values, and attitudes towards love and marriage. Understanding the fluctuations in marriage and divorce rates can provide insights into how these changes might influence the themes, narratives, and ultimately, the financial success of romantic comedies.

Several studies have explored the relationship between societal trends, including marriage and divorce rates, and their impact on media content. For instance, Banjo suggests that media consumption, especially of romanticized content, can foster unrealistic expectations and perceptions about romantic relationships, potentially influencing individuals' attitudes and behaviors in their own relationships<sup>[2]</sup>.

Ray highlights the concept of relational schemas, which are cognitive representations of how individuals expect relationships to function. The influence of romantic movie media on the development of these schemas suggests that individuals with high exposure to such media may internalize and seek to replicate the relationship patterns depicted in films<sup>[3]</sup>.

Ash (1994) adds to this by stating that when individuals consume romantic comedies, they are often presented with idealized depictions of romantic partners who effortlessly embody multiple roles and fulfill all relationship needs. This can further perpetuate unrealistic expectations within viewers' own relationships, potentially leading to dissatisfaction or disillusionment when reality fails to meet the standards set by media portrayals, and eventually divorce<sup>[1]</sup>.

However, while these studies provide valuable insights into the individual components of romantic comedies and societal trends, there is a gap in research specifically examining the intersection between marriage and divorce rates and the financial success of romantic comedies over time.

In this study, we aim to fill this gap by analyzing the profitability of romantic comedies spanning from 1990 to 2022 in relation to fluctuations in marriage and divorce rates in the United States. By identifying keywords associated with romantic comedies and using box office data to measure profitability, we seek to uncover any discernible trends or correlations between societal relationship dynamics and the commercial success of romantic comedies. This analysis will contribute to a deeper understanding of how societal changes impact media consumption patterns and the financial performance of films within the romantic comedy genre.

### The Top "Love-Themed/Romantic" Movie Each Year During 1990-2022 by Budget and Revenue

For this analysis, we chose to show the relationship between movies with the keyword tag of 'love' to their respective budgets and revenue. We only wanted to see the top grossing films related to 'love' during 1990 to 2022 to correspond with our marriage and divorce datasets. We chose to use the keyword 'love' instead of the genre 'romance' due to the major overlap in movies that were not related to 'romantic' movies but rather movies with 'romantic' elements in it. We wanted to focus on analyzing movies that were 'love' centered which is why we condensed our dataset to fit this specific category. 

We chose to pull out each movie's budget and revenue to show the financial incentive movie studios have when deciding to make 'love' themed movies. We also decided to use a bar graph to compare the budget to the revenue to show how financially successful each movie was during 1990-2022. 

```python
movies = pd.read_csv('tmdb_5000_movies.csv')


love = movies[movies['keywords'].str.contains('love', case=False, na=False)].copy()
love['release_year'] = pd.to_datetime(love['release_date'], errors='coerce').dt.year
love = love[(love['release_year'] >= 1990) & (love['release_year'] <= 2022)]
love = love.dropna(subset=['budget', 'revenue'])
love = love[(love['budget'] != 0) & (love['revenue'] != 0)]
love['budget'] = love['budget'].replace('[\$,]', '', regex=True).astype(float)
love['revenue'] = love['revenue'].replace('[\$,]', '', regex=True).astype(float)

top_lovemovies = love.loc[love.groupby('release_year')['budget'].idxmax()]

plt.rcParams.update({'font.size': 30})
plt.figure(figsize=(30, 30))
plt.rcParams.update({
    'font.size': 30,
    'axes.titlesize': 40,
    'axes.labelsize': 30,
    'xtick.labelsize': 30,
    'ytick.labelsize': 30,
    'legend.fontsize': 30
})

bar_width = 0.4
plt.bar(top_lovemovies['release_year'] - bar_width / 2, top_lovemovies['budget'], bar_width, color='plum', label='Budget')
plt.bar(top_lovemovies['release_year'] + bar_width / 2, top_lovemovies['revenue'], bar_width, color='hotpink', label='Revenue')
plt.xlabel('Year')
plt.ylabel('Amount on Dollars (In Billions)')
plt.title('The Top "Love-Themed/Romantic" Movie Each Year During 1990-2022 by Budget and Revenue')

for i, row in top_lovemovies.iterrows():
    plt.text(row['release_year'] - bar_width / 2, row['budget'], row['title'], ha='center', va='bottom', rotation=90)


plt.xticks(top_lovemovies['release_year'], rotation=45)

plt.legend()
plt.tight_layout()
plt.show()
```

From 1990-1999, James Cameron's 1997 movie "Titanic" was able to make 10x their original budget grossing over 2.2 billion dollars in revenue with a budget of 200 million dollars. "Titanic" was an extreme outlier when it came to 'love' themed movies in this decade. The second highest earning film was Disney's 1994 movie "The Lion King" grossing over 968 million dollars with a budget of 45 million dollars. 

From 2000-2009 we saw more 'action' based films with elements of 'romance' and 'love' rather than films that were solely made with the intention of romance being the central theme. The 2007 movie "Pirates of the Caribbean: At World's End" had the highest revenue with "Harry Potter and the Half Blood Price" in 2009 following closely behind. This decade seemed to have movies with a higher budget compared to the 90s and 2010 onwards.

From 2010-2019, there was a lull in revenue related to 'love' themed movies as four out of ten movies were not able to make back their original budget. The movies "R.I.P.D" (2013), "Aloha" (2015), "Live by Night" (2016), and "The Space Between Us" (2017) all were not able to break even as the movie studios lost money rather than gain a profit for producing these films. 

From 2020-2022, the lowest grossing film was in 2020 with Thomas Bezucha's "Let Him Go". There could have been many different factors in why this movie wasn't able to make back its original budget as this was the year a global pandemic affected the world as well as film industry. Bouncing back in 2021 and 2022, with "Spiderman: No Way Home" in 2021 grossing 1.9 billion dollars and "Avatar: The Way of Water" in 2022 grossing 2.3 billion dollars. Both of these movies were sequels which could have also attributed to this increase in revenue and audience reception. 

Overall this graph shows that movies with ‘love’ themes are able to make the movie studios money even if the budget is over 100 million dollars and that movies that were the most financially lucrative were movies that were ‘action’ based with ‘love’ elements. Outliers such as “Titanic” and “Avatar: The Way of Water” shows that movie goers are willing to watch movies that are ‘romance’ based with an added element of ‘action’.

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import ast
```

```python
df = pd.read_csv('TMDB_movie_dataset_v11.csv')

def extract_keywords(keywords_str):
    keywords_list = ast.literal_eval(keywords_str)
    return " ".join([keyword['name'] for keyword in keywords_list])

df['keywords_extracted'] = df['keywords'].apply(extract_keywords)

all_keywords = " ".join(df['keywords_extracted'])

wordcloud = WordCloud(width=800, height=400, background_color='white',).generate(all_keywords)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### Profitability

```python
popular_lovemovies['Profit Ratio'] = popular_lovemovies['revenue']/popular_lovemovies['budget']
popular_lovemovies['Profitable'] = (popular_lovemovies['Profit Ratio'] >= 1.20).astype(float)
popular_lovemovies = popular_lovemovies[['budget', 'revenue','original_title', 'release_year', 'Profit Ratio', 'Profitable']]
popular_lovemovies.rename(columns={'budget': 'Budget', 'revenue': 'Revenue', 'original_title': 'Title', 'release_year': 'Year' }, inplace=True)
popular_lovemovies
```

In the Profitable column we can see that 1.0 means the movie is profitable and meets the 20% profit margin threshold. On the other hand, 0.0 means the movie is not profitable. 

```python
annual_profitability = popular_lovemovies.groupby('Year')['Profitable'].mean().reset_index()
annual_profitability = pd.merge(popular_lovemovies[['Title', 'Year']], annual_profitability, on='Year')
annual_profitability
```

The "annual_profitability" dataframe tells us if the popular love/romance movies for each year are profitable or not.
